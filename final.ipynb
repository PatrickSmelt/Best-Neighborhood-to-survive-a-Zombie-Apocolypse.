# Metric : Population

For the analysis of the best Pittsburgh neighborhood for being able to survive a zombie attack, one of the datasets we chose to analyze was population of the  neighborhoods. The reason is if there is a lot of people there will be more zombies to turn.

```python
import pandas as pd
import geopandas
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
hoods = pd.read_csv("hoods.csv", sep=",")
hoods
```

```python
myhoods = geopandas.read_file("Neighborhoods_.shp")
myhoods.plot()
```

```python
population = pd.read_csv("total-population.csv", sep=",")
population
```

```python
mergepopulation = myhoods.merge(population, how='left', left_on='hood', right_on='Neighborhood')
mergepopulation
```

```python
mergepopulation.plot()
```

```python
mergepopulation.dropna()
mergepopulation.plot(column='Estimate; Total', legend= True)
```

```python
# import dataset
people = pd.read_csv("total-population.csv")
# filter to important info
num_people = people.groupby("Neighborhood").sum()['Estimate; Total']
num_people.sort_values(ascending=False)
```

```python
people = people.sort_values('Estimate; Total', ascending=False)
people.tail(20).plot.barh(x='Neighborhood')
```

So in my opinion the 'best' neighborhood is Chateau. It has the least population. It will survive the zombie attack.
