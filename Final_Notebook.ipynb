```python
#First Input Pandas to allow us to make some data visualizations
import pandas as pd
import geopandas
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Determining the Best Neighborhood to live in Pittsburgh to survive a zombie apocolypse.
#Starting off by creating Dictionary

treesFile = pd.read_csv("trees-pitt.csv")

trees_in_neighborhood = dict()

#caluate frequency of trees in each neighborhood and see how they provide the best coverage.
for index, row in treesFile.iterrows():
    trees = row['height']
    
    #assigning trees to the row of height
    
    if trees > 0:
        #telling the code that if the heigh is above 0 it can be classified as a tree.
        #the more trees their are, the more coverage you have which makes it harder for zombies.
        neighborhood = row['neighborhood']
        #variable assinged to neighborhood for its row
        if (neighborhood not in trees_in_neighborhood):
            trees_in_neighborhood[neighborhood] = 1
        else:
            trees_in_neighborhood[neighborhood] += 1
            
            #telling the dictionary to add up the number of trees in each neighborhood.

            
print(trees_in_neighborhood)







        
```
